```
import matplotlib.pyplot as plt
```

```
def log_map(x, r):
  return r * x * (1 - x)
```

```
def find_ep(x, x_):
  eps = 0.001
  position = -1
  for i in range(len(x)):
    if abs(x_ - x[i]) < eps:
      position = i

  return position
```

```
def iterate(x, r, candidates, positions):
  x_ = log_map(x, r)
  candidates.append(x_)

  iter = 1
  while (iter < 500):
    x = x_
    x_ = log_map(x, r)
    
    does_find = find_ep(candidates, x_)
    if (does_find < 1):
      candidates.append(x_)
    else:
      positions.append(does_find)

    iter += 1

  return candidates, positions
```

```
x = 0.01
r = 0

parameters = []
points = []

while (r < 4):
  candidates = []
  positions = []

  candidates, positions = iterate(x, r, candidates, positions)

  for i in range(len(positions)):
    points.append(candidates[positions[i]])
    parameters.append(r)

  r = r + 0.01
  
```

```
plt.plot(parameters, points, '.', ms = 0.5, color = 'teal')
plt.show()
```
